# Data re-uploading for a single-qubit universal classifier - Vol. I

## A new model for Quantum Machine Learning with one or few qubits


### Abstract
Welcome, qiskitters. Today we will learn a little bit about Quantum Machine Learning and how to apply it using the IBMQ-experience platform. We will solve toy-problems related to classification in Unsupervised Machine Learning. One qubit has got enough computational power for classifying classical data. The critical idea that is the core of this method is *re-uploading*. Re-uploading classical data several times into a single-qubit quantum circuit via unitary operations allows us to classify any kind of data into any number of classes. This idea can also be extended to more qubits for improving the performance of the classifier. This tutorial is based on the work in [1].

### Introduction

Machine Learning is one of the most useful applications found in classical computation, and, in recent years, it has been developed and implemented in many different circumstances. Nowadays, it is used broadly for solving many different types of problems. The main branch of Machine Learning, where we will focus, is known as Supervised Machine Learning. This paradigm consists in classifying data from a given dataset[1]. Let us imagine we have got a large set of data, which we know is split into two (or more) different categories. We want to develop a machine which is able to distinguish between these classes. The standard way of achieving it is by letting the machine learn the features of the dataset that are useful for telling points apart. We show a *training set* to the algorithm, i. e.: we give pairs of point and class to the classifier, and it learns which properties are important for determining the class of every point. Then, we show a *test set* to the classifier with no classes, and it has to decide to which class belongs each point in the test set. If the success rate is high, then the classifier has learnt the global features of the dataset and will succeed in the task of classifying any point. The training of the machine is critical for the accomplishment of this goal. In this tutorial, every time we mention Machine Learning, we take for granted that it is supervised.

Machine Learning is also one of the most promising areas in the field of Quantum Computing. Recently, many proposals have been published, and there are many different strategies for merging both fields in a common paradigm called Quantum Machine Learning. Some QML algorithms take advantage of the computational capabilities of Quantum Computing for performing classical tasks, as in Ref. [[3](https://arxiv.org/abs/1806.11463)]. Some other approaches tackle a Machine Learning problem in quantum ways. These methods can still be inspired in classical classification algorithms [4,5] or make use of quantum parameterized circuits [6,7]. This list of references does not aim to be complete, but a small approach to previous work. For interested qiskitters, we encourage you to read the book in [8]. Up to now, there is no agreement on how to build a model for Quantum Machine Learning performing as classical models. 

One fundamental limitation of Quantum Mechanics, and thus of quantum computing, is the *no-cloning* theorem [9]. In general, it is simply not possible to copy some arbitrary data encoded in a quantum state into another quantum system. If there were a classical no cloning theorem, Machine Learning as we know it would not exist, as far as data must be re-input several times for computing different quantities during the processing. For instance, neural networks, which are a simple and very used model of Machine Learning algorithms, perform series of calculations with the same input. Therefore, QML could only mimic the tactics of its classical conterpart if there are many copies of quantum data, or if classical data is introduced several times in the quantum circuit. As quantum memories (qRAMs) are not available yet [10], we propose an algorithm of QML following the second estrategy.

We present now a QML model that makes use of quantum parameterized circuits for classifying data following the recipe in [1]. The minimal needs of the circuit are only one qubit and few gates. Circuits with only one qubit have no quantum advantage, as they are easy to simulate, but may be useful as part of larger circuits. This method must be extended to larger and entangled quantum systems in order to get a quantum advantage for QML-only purposes. 

In this model, we start with a qubit in the ground state, and then apply several gates on it. The gates receive as input the data we want to classify, that change with every point in the training set, and some inner parameters that are common to all points. This way the building-block, uploading and processing is repeated several times in a row. The processing of this data is performed in different steps sequentially, and this is a key difference to neural networks, where data is processed in parallel, see Figs. 1 and 2.

### Structure of a single-qubit classifier
Let us take the dataset we want to classify as 
$$ \mathcal{D} = \lbrace \vec x, y \rbrace $$
where $\vec x$ is the set of input data and $y$ is the label of each $\vec x$. We have to know since the begining of the problem the total number of different $y$ in the problem. 

We will give now a general idea of the procedure. We want to create a unitary operation depending on $\vec x$ and some optimizable parameters $\Phi$, such that $$ U(\Phi, \vec x)| 0 \rangle = |\psi(\Phi, \vec x)\rangle .$$
On the other hand, the labels $y$ will be mapped into quantum states $$ y \longrightarrow |\psi_y\rangle. $$Both quantum states should be similar $$ |\psi(\Phi, \vec x)\rangle \sim |\psi_y\rangle $$ for every pair $(\vec x, y)$. 


Following the scheme from Fig. 2, we can describe the output quantum state as the result of the circuit in Fig. 3. The gates $U(\vec x)$ introduce the classical input into the circuit, while the gates $U(\vec{\phi_i})$ process that information. The parameters $\vec{\phi_i}$ are contained into the quantity $\Phi$. Notice that there are $N$ different $\vec{\phi_i}$. In addition, as two different rotations compose another one, we can rearrange $U(\vec x)$ and $U(\vec{\phi_i})$ into one common gate that will be

$$
U(\vec{\phi_i}, \vec x) = U3\left(\theta(\vec{\phi}_i, \vec x), \varphi(\vec{\phi}_i, \vec x), \lambda(\vec{\phi}_i, \vec x)\right),
$$

where $U3$ is the usual gate from *IBMQ*

$$
U3(\theta, \varphi, \lambda) = \begin{pmatrix}
\cos(\theta / 2) & -\sin(\theta / 2) e^{i\lambda} \\ \sin(\theta / 2) e^{i\varphi} & \cos(\theta/2) e^{i\lambda} e^{i\varphi}
\end{pmatrix}.
$$

After applying many different $L(i)$ gates the final result will be

$$
|\psi(\Phi, \vec x)\rangle = \prod_{i = 0}^L U(\vec{\phi_i}, \vec x) | 0 \rangle$$

In this tutorial we will focus on classifying points in a plane. Thus, every unitary gate will have 5 parameters $(\phi_0, \ldots, \phi_4)$ to optimize. The explicit form of the gates is

$$
U3(\theta, \phi, \lambda) \rightarrow U3(\phi_0 x_0 + \phi_1, \phi_2 x_1 + \phi_3, \phi_4)
$$

It is important to mention that the dimension of input data $\vec x$ is not a limiting factor for this problem. If $x$ has got dimension up to $3$, all the information can be encoded in one unitary gate. On the contrary, if this dimension is larger, we can just apply more than one gate per input-step.

<img src="reuploading_circuit.png" width="500"> 
*Fig. 3: Circuit for the single-qubit classifier model. Every $U$ is a general unitary gate $U3$ depending on some parameters. The gates $U(\vec x)$ introduce the classical input into the circuit, while the gates $U(\vec{\phi_i})$ process that information. Two gates are rearranged to become a gate $U(\vec{\phi_i}, \vec x)$*
<br>


### Measurement and optimizing 
As the parameters $\Phi$ of the classifier must be optimizes, we need to generate a cost function which drives this optimization. The general idea of the solving proposal suggests to use the fidelity between output and target state as the figure of merit. We expect the output states for some $\vec x$ to be close to $| \psi_y\rangle$ for every pair $\lbrace \vec x, y\rbrace$. With that in mind, we choose the cost function

$$
\chi^2(\Phi) = \frac{1}{M}\sum_{\mu = 1}^M (1 - \mathcal{F}(\psi(\Phi, \vec x_\mu), \psi_{y_\mu}))^2
$$

where $ \mathcal{F}(\psi(\Phi, \vec x_\mu), \psi_{y_\mu}) = |\langle \psi_{y_\mu}|\psi(\Phi, \vec x_\mu)\rangle|^2$, and $\mu$ runs over all samples in the dataset. This fidelity can be intuitively understood for a binary dataset as some kind of probability of belonging to one particular class. There is an obvious balance between both choices: the closer an state is to $| 0 \rangle$, the further it is from $| 1 \rangle$.

#### Optimization
In this tutorial, the process of optimizing this cost function will be performed using the `qasm simulator` in order to be faster. Many calculations are required with this purpose, so the usage of actual quantum devices is prohibitevely expensive. However, any user can change this choice to quantum devices by adjusting few parameters in the code.

With regard to the minimization method, there are lots of them. In this tutorial we let the Powell method implemented in the python package *scipy* do all the work. We found this is the method that best fits our purposes. *Scipy* is a very convenient library for many optimization problems. As a matter of fact, *IBMQ's aqua* uses also *scipy* as a subroutine for optimizing circuits.

Through this whole tutorial, the parameters obtained via optimization are provided. However, that does not mean that the these very parameters are the optimal ones. Different or more sophisticated minimization methods could return other sets of parameters that solve the problem better. However, at this point, a good approximation is sufficient for showing how the classification algorithm works.

#### SWAP test
We have to deal with the problem of measuring the fidelity of two states. The most direct way of doing it is the SWAP test.
<img src="swap_test.png" width="300"> 
*Fig. 4: Circuit for the swap test. $| \alpha \rangle$ and $| \beta \rangle$ are unknown states but it is possible to compute their relative fidelity using this circuit. The $H$ gates are Hadamard gates, and the entangling one is a Fredkin gate.*
<br>

This is a useful tool for measuring fidelities between states. Given two states $| \alpha \rangle$ and $| \beta \rangle$, the way of measuring their relative fidelity is by performing the circuit from Fig. 4. The fidelity can be directly computed with

$$
P_{0, 1} = \frac{|\langle \alpha | \beta \rangle |^2 \pm 1}{2} 
$$

where $P_i$ is the probability of measuring $i$ in the computational basis. 

We will spend some time developing the details on the swap test. In the following lines the normalization factors are omitted for convenience, but recovered at the end. We start with a quantum state
$$ |\psi\rangle = |0\rangle|\beta\rangle|\alpha\rangle $$
Then a Hadamard gate applies on the ancilla qubit, and we get
$$ |\psi\rangle = (|0\rangle + |1\rangle)|\beta\rangle|\alpha\rangle $$
and then apply a controlled swap gate
$$ |\psi\rangle = |0\rangle|\beta\rangle|\alpha\rangle + |1\rangle|\alpha\rangle|\beta\rangle $$
and another Hadamard gate
$$ |\psi\rangle = (|0\rangle + |1\rangle)|\beta\rangle|\alpha\rangle + (|0\rangle - |1\rangle)|\alpha\rangle|\beta\rangle = |0\rangle (|\beta\rangle|\alpha\rangle + |\alpha\rangle|\beta\rangle) + |1\rangle (|\beta\rangle|\alpha\rangle - |\alpha\rangle|\beta\rangle)$$
Thus, with the normalization factor, the probability of measuring the $|0\rangle$ state will be
$$ P_0 = \frac{1}{4}(\langle\beta|\langle\alpha| + \langle\alpha|\langle\beta|)(|\beta\rangle|\alpha\rangle + |\alpha\rangle|\beta\rangle) = \frac{|\langle \alpha | \beta \rangle |^2 + 1}{2}.$$
To sum up, an equivalent calculation can be done for measuring $|1\rangle$.

### The problem: Classifying points In and Outside a Circle

We will tackle now a very simple problem: let us imagine a squared area with a circle right in the middle. The circle has got a radius such that the area of the circle is half the total area of the square. We will build a *quantum* classifier that tells us whether a given point from this area is inside or outside the circle. 

Our training dataset will be a grid specified below. Yellow points are inside the circle and blue are outside. The labels are $0$ for *out* and $1$ for *in*. As it is a binary dataset, the obvious choice is to take the states of the computational basis $| 0 \rangle$ and $| 1 \rangle$. Roughly speaking, we want a cost function such that if the label is $0$ $(1)$, measuring the quantum state and getting $| 0 \rangle$ ($| 1 \rangle$) is more likely. 

```python
grid = 10
x = np.linspace(-.9,.9,grid)
X = list(itertools.product(x, x))
X_train = np.asfarray(X)
Y_train = circle_labels(X_train)

fig = draw_circle((X_train, Y_train), title='In and Out')
```

Now, it is time to create the circuit as a combination of adjustable unitary gates and the swap test. The $U$ gate is as explained before. There is another $X^y$ gate in a second qubit that will be needed for the swap test. This gate will be a $X$ gate only if the point belongs to the class $ 1 $, i. e.: it is inside the circle. Otherwise, it will be an identity gate and have no effect. For obvious reasons, this identity gate is not in the quantum circuit.
<img src="circuit_circle.png" width="400"> 
*Fig. 4: Circuit for the classification of a circle. The $X^y$ gate will have effect only if the point is inside the circle.*
<br>

In this cell below we define the circuit and another function for computing the fidelity with the $|0\rangle$ state. This will be useful later. We define also the cost function for this problem.

```python
def measurement_circle(x, y, parameters, device, err_like_dev, noise_objects):
    """
    Create the basic circuit for this problem
    Input:
        x: the x input of the data
        y: the label of the data
        parameters: variables of the classifier
        device: name of the IBMQ device in which we want to run the circuit, it can be a qasm simulator or a real device
        err_like_dev: which device we want to take as noise model
                It will only work if device is the qasm simulator, otherwise it has no effect
        noise_objects: noise model, coupling map and basis gates for the error model
                
    Output:
        The fidelity between the state given by (x, y) and its corresponding label state
    
    Notice that the circuit is designed in a way that fits the architecture of 'ibmqx2'
    """
    if len(parameters) % 5 != 0:
        raise ValueError('The number of parameters is not correct')
        # Error message in case the number of parameters does not fit the problem
        
    layers = len(parameters) // 5
    parameters = parameters.reshape(layers, 5) # It is easier to manage variables this way
    
    C = QuantumCircuit(3,1) # We create the Quantum circuit, we need 3 qubits and one classical register
    
    if y == 1:
        C.x(1) # This gate will appear only if the point is inside the circle
        
    C.h(2) # This is part of the SWAP test
    
    for l in range(layers): 
        C.u3(parameters[l,0] * x[0] + parameters[l, 1], parameters[l, 2] * x[1] + parameters[l, 3], parameters[l, 4], 0)
        # This loop creates all unitary gates for classifying data, there is one gate per layer
        
    C.cswap(2, 0, 1) # This is part of the SWAP test
    C.h(2) # This is part of the SWAP test
    C.measure([2], [0])  # This is part of the SWAP test, will give us the fidelity
    
    try:
        backend = Aer.get_backend(device)# Try to select the QasmSimulator from the Aer provider
        
    except: # If the device given is not the qasm simulator
        provider = IBMQ.get_provider(hub='ibm-q') 
        backend=provider.get_backend(device)  # This selects the real devices, one must be logged into IBMQ for using the function
        C=transpile(C, backend=backend, optimization_level=1) # Transpiling of the circuit, this adjusts the circuit to be convenient for the chosen provider, the optimization level can be adjusted for better results

    sh = 1024 # Number of shots for the execution, this can be easily changed
    if err_like_dev!= None: # If we want a noise model, other functions are designed to make this work even with no input variables in the function
        noise_model, coupling_map, basis_gates = noise_objects # Unpack the noise objects, this way we do not have to compute them every time
        result = execute(C, backend, shots=sh, noise_model=noise_model,
                       coupling_map=coupling_map,
                       basis_gates=basis_gates).result() # Execution
    else: # If we consider the device as fault tolerant
        result = execute(C, backend, shots=sh, optimization_level = 1).result() # Execution
    counts = result.get_counts(C) # Get the number of counts for each state
    fidelity = 2 * counts['0'] / sh - 1 # Compute the fidelity 
    return fidelity


def check_circle(X, parameters, device, err_like_dev=None):
    """
    Compute the guessed label of a X dataset, this function needs no labels
    Input:
        X: the x input of the dataset
        parameters: variables of the classifier, which at this point should be optimized for good results
        device: name of the IBMQ device in which we want to run the circuit, it can be a qasm simulator or a real device
        err_like_dev: which device we want to take as noise mode
                It will only work if device is the qasm simulator, otherwise it has no effect
                
    Output:
        The fidelity between the state given by (x, y) and its corresponding label state
    """
    noise_objects=None # This section creates the noise model in case we want it
    if err_like_dev!= None:
        print('Checking with noise model')
        provider = IBMQ.get_provider(hub='ibm-q')
        err_device = provider.get_backend(err_like_dev)
        properties = err_device.properties()
        coupling_map = err_device.configuration().coupling_map
        noise_model = noise.device.basic_device_noise_model(properties)
        basis_gates = noise_model.basis_gates
        noise_objects = noise_model, coupling_map, basis_gates
        
    Fidelities = np.empty(len(X)) # Empty vector for all fidelities
    for i,x in enumerate(X):
         Fidelities[i]= measurement_circle(x, 1, parameters, device, err_like_dev, noise_objects) # Checks if it is inside the circle 

    return Fidelities
    
    
def cost_function_circle(parameters, Data, device, err_like_dev, noise_objects):
    """
    Compute the cost function for a dataset
    Input:
        Data = (X, Y): the x and y input of the dataset
        parameters: variables of the classifier, which at this point should be optimized for good results
        device: name of the IBMQ device in which we want to run the circuit, it can be a qasm simulator or a real device
        err_like_dev: which device we want to take as noise mode
                It will only work if device is the qasm simulator, otherwise it has no effect
        noise_objects: noise model, coupling map and basis gates for the error model
                
    Output:
        The cost function for a training Dataset
    """
    noise_model, coupling_map, basis_gates = noise_objects
    X = Data[0]
    Y = Data[1]
    chi2 = 0
    for x, y in zip(X, Y):
        chi2 += (1 - measurement_circle(x, y, parameters, device, err_like_dev, noise_objects))**2 # Compute the cost function for every point and sum them all
        
    return chi2 / len(Y)

def get_optimal_parameters(parameters, Data, device, err_like_dev=None):
    """
    Get the optimal parameters of a Dataset
    Input:
        Data = (X, Y): the x and y input of the dataset
        parameters: variables of the classifier. At this point they should be optimized for good results
        device: name of the IBMQ device in which we want to run the circuit, it can be a qasm simulator or a real device
        err_like_dev: which device we want to take as noise mode. 
                It will only work if device is the qasm simulator, otherwise it has no effect
                
    Output:
        Optimal parameters
    """
    noise_objects=(None, None, None) # This section creates the noise model in case we want it.
    if err_like_dev!= None:
        print('Optimization with noise model')
        provider = IBMQ.get_provider(hub='ibm-q')
        err_device = provider.get_backend(err_like_dev)
        properties = err_device.properties()
        coupling_map = err_device.configuration().coupling_map
        noise_model = noise.device.basic_device_noise_model(properties)
        basis_gates = noise_model.basis_gates
        noise_objects = noise_model, coupling_map, basis_gates
        
    result_min = minimize(cost_function_circle, parameters, args=((X_train, Y_train), simulator, err_like_dev, noise_objects), method='Powell')
    
    return result_min
```

Let us see now how this algorithm performs. We use the optimal parameters obtained before in the circuit and measure the relative fidelity between the output state and $| 0 \rangle$ in the ancilla qubit. The fidelity we obtain can be interpreted as the probability of every point for being inside or outside the circle.

```python
F_train = check_circle(X_train, optimal_parameters, simulator) # The err_like_dev is optional
fig=draw_circle((X_train, F_train), colorbar=True, title = 'Probability of being inside the circle')
```

We see that points at the center of the area are closer to $1$ than points at the outer part. However, at the end of the day we want a prediction for a point to be *IN* or *OUT*. We have to set a way to decide to which class a point belongs. As a first approach, we can apply a rule of maximum: if measuring $0 (1) $ is the most probable situation, then the point belongs to class $0 (1)$. The result we obtain using this method is depicted below. It is important to mention that this may be not the optimal rule.

```python
Z_train = np.round(F_train)

checks_train = (Z_train == Y_train).astype(int)
fig=draw_circle((X_train, Z_train), check=False, title='In and Out')
fig=draw_circle((X_train, checks_train), check=True, title='Right and wrong')
print('The classifier guesses the right label a {} % of times'.format(100* sum(checks_train) // len(checks_train)))
```

We are now about to test our classifier against unseen data. This is crucial to check whether our classifier learns the general features of the data or, on the contrary, learns only the training dataset. As a first step, we need to compute the fidelities and test them. Then, we decide the class of every point with the rule of maximum.

```python
test=100
X_test = 1.8*np.random.rand(test, 2) - .9
Y_test = circle_labels(X_test)

F_test = check_circle(X_test, optimal_parameters, simulator)

fig=draw_circle((X_test, F_test), colorbar=True, title = 'Probability of being inside the circle')

Z_test = np.round(F_test)

checks_test = (Z_test == Y_test).astype(int)
fig=draw_circle((X_test, Z_test), check=False, title='In and Out')
fig=draw_circle((X_test, checks_test), check=True, title='Right and Wrong')

print('The classifier guesses the right label a {} % of times'.format(100* sum(checks_test) // len(checks_test)))
```

Let us now design a better rule. Instead of applying the rule of maximum, we will set a bias $p$, i.e.: if the probability of measuring $| 0 \rangle$ does not exceed $p$, the point belongs to class $0$. The optimal value $p$ can be computed by running over all possible different values of it. This operation will always improve the results of the classifier as we can always set $p = 1/2$ and recover the rule of maximum.

```python
s_test, p_test = swipe_check(F_test, Y_test)

Z_test = (F_test > p_test).astype(int)
checks_test = (Z_test == Y_test).astype(int)
fig=draw_circle((X_test, Z_test), check=False, title='In and Out')
fig=draw_circle((X_test, checks_test), check=True, title='Right and Wrong')

print('The classifier guesses the right label a {} % of times'.format(100 * sum(checks_test) // len(checks_test)))
```

```python
qtest=100
X_qtest = 1.8*np.random.rand(qtest, 2) - .9
Y_qtest = circle_labels(X_qtest)

F_qtest = check_circle(X_qtest, optimal_parameters, simulator, err_like_dev=quantum_device) # For a noisy simulation
#F_qtest = check_circle(X_qtest, optimal_parameters, quantum_device) # For an actual quantum computer

#fig=draw_circle((X_qtest, F_qtest), colorbar=True)

Z_qtest = (F_qtest > p_test).astype(int)
checks_qtest = (Z_qtest == Y_qtest).astype(int)
fig=draw_circle((X_qtest, Z_qtest), check=False)
fig=draw_circle((X_qtest, checks_qtest), check=True, title='Right and Wrong')

print('The classifier guesses the right label a {} % of times'.format(100* sum(checks_qtest) // len(checks_qtest)))
```

## Conclusions

We have learnt by solving this problem that one qubit has got enough computational capability as to perform a binary classification. We must help it with classical optimizers though. Results are pretty good, over a 
90% of success for few parameters. We could say that when connection between the quantum computer and our laptops improves, or when an optimizer for quantum computers is created, solving this problem will be an easy task for any quantum machine.

However, during the explanation at the very beginning of this tutorial, no theoretical limits were found for this method. Indeed, no theoretical limits have been yet found, see Ref. [1]. The questions we should ask ourselves are the following: Can we classify data with more than two labels? Could we use more qubits?

We can show that non-binary datasets can be classified and more qubits can be used with this method. This is explored in the `jupyter notebook`s Vols. II and III of this tutorial.
